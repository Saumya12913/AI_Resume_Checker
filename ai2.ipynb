```python
import gradio as gr
import pdfplumber
import re
```

```python
def evaluate_resume(file):

    # Extract text from uploaded PDF
    with pdfplumber.open(file.name) as pdf:
        resume_text = ""
        for page in pdf.pages:
            resume_text += page.extract_text()

    score = 0
    feedback = []

    if len(resume_text.split()) > 150:
        score += 20
        feedback.append("вњ… Good resume length.")
    else:
        feedback.append("вќЊ Resume is too short.")

    skills_keywords = ["python", "java", "sql", "machine learning", "project"]
    found_skills = [skill for skill in skills_keywords if skill in resume_text.lower()]

    if found_skills:
        score += 30
        feedback.append(f"вњ… Skills detected: {', '.join(found_skills)}")
    else:
        feedback.append("вќЊ No strong skills detected.")

    if re.search(r"\S+@\S+\.\S+", resume_text):
        score += 10
        feedback.append("вњ… Email found.")
    else:
        feedback.append("вќЊ Email not found.")

    if "experience" in resume_text.lower():
        score += 20
        feedback.append("вњ… Experience section detected.")
    else:
        feedback.append("вќЊ Add Experience section.")

    if "education" in resume_text.lower():
        score += 20
        feedback.append("вњ… Education section detected.")
    else:
        feedback.append("вќЊ Add Education section.")

    final_score = min(score, 100)

    return f"Resume Score: {final_score}/100\n\n" + "\n".join(feedback)
```

```python
app = gr.Interface(
    fn=evaluate_resume,
    inputs=gr.File(label="Upload your Resume (PDF only)"),
    outputs="text",
    title="AI Resume Checker",
    description="Upload your resume PDF and get instant evaluation."
)
```

```python
app.launch(share=True)
```
